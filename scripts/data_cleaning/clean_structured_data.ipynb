# World Governance Indicators

```python
df = pd.read_csv('/content/WGI_1996_2023.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Series Code'].unique()
```

```python
df_filtered = df.loc[df['Series Code'].str.contains(r'.*\.EST', na=False)]
```

```python
df_filtered.head()
```

```python
year_cols = [col for col in df_filtered.columns if '[YR' in col]
id_vars = [col for col in df_filtered.columns if col not in year_cols]

df_melted = pd.melt(df_filtered, id_vars=id_vars, value_vars=year_cols, var_name='year_col', value_name='value')

# Extract just the year number from the year_col and convert to integer
df_melted['year'] = df_melted['year_col'].str.extract(r'(\d{4})').astype(int)

# Convert the 'value' column to numeric, coercing errors to NaN
df_melted['value'] = pd.to_numeric(df_melted['value'], errors='coerce')

# Drop the original 'year_col' as we have the cleaned 'year' column
df_melted = df_melted.drop(columns=['year_col'])

display(df_melted.head())
```

```python
df_melted['Series Name'].unique()
```

```python
df_melted['Series Name'] = df_melted['Series Name'].apply(lambda x: x.split(':')[0].strip().lower().replace(' ', '_'))
display(df_melted.head())
```

```python
df_melted = df_melted.loc[lambda x: x.year == 2023].reset_index().drop(columns = ['Series Code','index'])
```

```python
df_melted
```

```python
df_melted = df_melted.loc[lambda x: x['Country Code'].isin(mapping.values())]
```

```python
df_melted.head()
```

```python
df_pivot = df_melted.pivot_table(index=['Country Name', 'Country Code', 'year'], columns='Series Name', values='value').reset_index()

df_pivot.columns.name = None # Remove the 'Series Name' column name
display(df_pivot.head())
```

```python
df_pivot.info()
```

```python
df_pivot.rename(columns = {'Country Code': 'country_code'}).drop('Country Name', axis=1).to_csv('WGI_Select_Countries.csv', index=False)
```

```python
df = pd.read_excel('/content/tariff_database_2025.xlsx')
```

```python
df.info()
```

```python
df.head()
```

```python
df['hts8'] = df['hts8'].astype(str)

valid_chapters = ('84', '85', '90')

mask_chapter = df['hts8'].str.startswith(valid_chapters)

df_elec = df[mask_chapter].copy()
```

```python
include_keywords = [
    "semiconductor", "integrated circuit", "IC", "microprocessor", "microcontroller",
    "logic", "memory", "flash", "DRAM", "SRAM", "EPROM", "EEPROM", "MPU", "MCU",
    "diode", "transistor", "thyristor", "LED", "capacitor", "resistor", "inductor",
    "transformer", "connector", "relay", "sensor", "oscillator", "crystal",
    "printed circuit", "PCB", "circuit assembly", "electronic component"
]

exclude_keywords = [
    "television", "smartphone", "washing machine", "refrigerator", "camera",
    "microwave", "printer", "monitor", "projector", "radio", "speaker", "headphone"
]

mask_include = df_elec['brief_description'].str.contains('|'.join(include_keywords), case=False, na=False)
mask_exclude = df_elec['brief_description'].str.contains('|'.join(exclude_keywords), case=False, na=False)

df_filtered = df_elec[mask_include & ~mask_exclude]
```

```python
df_filtered.info()
```

```python
df_filtered.shape
```

```python
df_filtered.head()
```

```python
include_keywords = [
    r"\bsemiconductor(s)?\b",
    r"\bintegrated circuit(s)?\b",
    r"\bmicroprocessor(s)?\b",
    r"\bmicrocontroller(s)?\b",
    r"\bdiode(s)?\b",
    r"\btransistor(s)?\b",
    r"\bthyristor(s)?\b",
    r"\bLED(s)?\b",
    r"\bcapacitor(s)?\b",
    r"\bresistor(s)?\b",
    r"\binductor(s)?\b",
    r"\btransformer(s)?\b",
    r"\bconnector(s)?\b",
    r"\brelay(s)?\b",
    r"\bsensor(s)?\b",
    r"\boscillator(s)?\b",
    r"\bcrystal(s)?\b",
    r"\bprinted circuit\b",
    r"\bPCB(s)?\b",
    r"\bcircuit assembly\b",
    r"\belectronic component(s)?\b"
]

exclude_keywords = [
    r"\btelevision(s)?\b",
    r"\bsmartphone(s)?\b",
    r"\bwashing machine(s)?\b",
    r"\brefrigerator(s)?\b",
    r"\bcamera(s)?\b",
    r"\bmicrowave(s)?\b",
    r"\bprinter(s)?\b",
    r"\bmonitor(s)?\b",
    r"\bprojector(s)?\b",
    r"\bradio(s)?\b",
    r"\bspeaker(s)?\b",
    r"\bheadphone(s)?\b"
]

inc_pattern = "|".join(include_keywords)
exc_pattern = "|".join(exclude_keywords)

mask_include = df_elec['brief_description'].str.contains(inc_pattern, case=False, na=False, regex=True)
mask_exclude = df_elec['brief_description'].str.contains(exc_pattern, case=False, na=False, regex=True)

df_filtered = df_elec[mask_include & ~mask_exclude].copy()
```

```python
df_filtered.head()
```

```python
exclude_keywords_extra = [ r"\bwaste\b", r"\boil(s)?\b", r"\bPCB(s)?\b", # chemical PCBs, not printed circuit boards
                          r"\bPCT(s)?\b", r"\bPBB(s)?\b", r"\bpolychlorinated\b", r"\bhalogenated\b" ]
```

```python
exclude_keywords_all = exclude_keywords + exclude_keywords_extra

exc_pattern = "|".join(exclude_keywords_all)

mask_exclude = df_elec['brief_description'].str.contains(exc_pattern, case=False, na=False, regex=True)

df_filtered2 = df_elec[mask_include & ~mask_exclude]
```

# World Bank Commodity Prices

## Monthly

```python
df = pd.read_excel('/content/Monthly-2.xlsx', sheet_name='Monthly Prices', header=4)
```

```python
df.head()
```

```python
second_header = df.iloc[0]
```

```python
second_header
```

```python
df.columns = [
    f"{col} {second_header[i]}" if pd.notna(second_header[i]) else col
    for i, col in enumerate(df.columns)
]
```

```python
df = df.iloc[1:].reset_index(drop=True)
```

```python
df.head()
```

```python
df_cleaned = df.dropna(subset=['Unnamed: 0']).copy()
df_cleaned = df_cleaned.assign(
    date = lambda x: pd.to_datetime(x['Unnamed: 0'].astype(str).str.replace('M','-', regex=False) + '-01')
)
display(df_cleaned.head())
```

```python
cols_to_keep = [
    col for col in df.columns
    if any(keyword in col for keyword in [
        "Crude oil",
        "Natural gas",
        "Copper",
        "Aluminum"
    ])
]

df_filtered = df_cleaned[["date"] + cols_to_keep]
```

```python
df_filtered.head()
```

```python
df_filtered = df_filtered.assign(
    year = lambda x: x.date.dt.year,
    month = lambda x: x.date.dt.month,
)
```

```python
df_filtered.head()
```

```python
df_melted = df_filtered.melt(
    id_vars = ['date', 'year', 'month'],
    var_name = 'commodity',
    value_name = 'price'
)
```

```python
df_melted['source'] = 'World Bank Pink Sheet'
```

```python
df_melted["commodity_clean"] = (
    df_melted["commodity"]
    .str.replace(r"\s*\(.*?\)", "", regex=True)
    .str.strip()
)
```

```python
df_melted.head()
```

```python
df_melted.rename({'price':'nominal_price'}, axis=1, inplace = True)
```

```python
df_melted = df_melted.assign(
    unit = lambda x: x.commodity.str.extract(r'\((.*?)\)')
).drop('commodity', axis=1)
```

```python
df_melted.head()
```

```python
df_melted.shape[0]
```

```python
df_melted.loc[~df_melted.commodity_clean.str.contains('average')]
```

```python
display(df_melted.head())
```

```python
df_melted[["commodity_family", "commodity_variant"]] = (
    df_melted["commodity_clean"]
    .str.split(",", n=1, expand=True)
)

df_melted["commodity_family"] = df_melted["commodity_family"].str.strip()
df_melted["commodity_variant"] = df_melted["commodity_variant"].str.strip()
```

```python
df_melted.head()
```

```python
df_melted = df_melted[['date', 'year', 'month', 'commodity_family', 'commodity_variant', 'unit', 'nominal_price', 'source']]
```

```python
df_melted.tail()
```

```python
df_melted.to_csv('monthly_world_bank_commodity_prices_1960_2025.csv', index=False)
```
