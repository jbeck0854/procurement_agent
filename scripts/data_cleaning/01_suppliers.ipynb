# Extracting Unique Country Codes


```python
unique_lpi_countries = set(lpi['country_code'].unique())
unique_port_countries = set(port['country_code'].unique())
unique_wgi_countries = set(wgi['country_code'].unique())
unique_combined_ppi_countries = set(combined_ppi['country_code'].unique())
# The 'tariff' DataFrame does not have a 'country_code' column, so it is excluded from this step.

all_unique_countries = unique_lpi_countries.union(
    unique_port_countries,
    unique_wgi_countries,
    unique_combined_ppi_countries
)

unique_countries = sorted(list(all_unique_countries))
print(f"Total unique countries identified: {len(unique_countries)}")
print(f"Unique countries: {unique_countries}")
```

# Assigning Countries to Relevance Tiers (based on manufacturing + export capability)

e.g., the US is mostly a CONSUMER, not a supplier.

```python
# Countries that dominate global semiconductor manufacturing and export capacity.
# Dominate supply
tier_1 = ['CHN', 'JPN', 'SGP', 'HKG', 'MYS', 'THA', 'DEU']
```

```python
# strong but not dominant
tier_2 = ['MEX', 'IND', 'CAN', 'FRA', 'GBR', 'AUS', 'IDN', 'NLD', 'USA']
```

```python
# low relevance in semiconductor manufacturing
tier_3 = [country for country in unique_countries if country not in tier_1 and country not in tier_2 and country != 'OAC']
```

# Mean Value of Relevant Metrics

```python
lpi_metrics = ['customs', 'infrastructure', 'international_shipments', 'logistics_competence', 'tracking', 'timeliness']
lpi_baseline = lpi.groupby('country_code')[lpi_metrics].mean()
print("LPI Baseline Metrics:")
print(lpi_baseline.head())
```

```python
port_metrics = ['median_days_in_port']
port_baseline = port.groupby('country_code')[port_metrics].mean()
print("Port Baseline Metrics:")
print(port_baseline.head())
```

```python
wgi_metrics = ['control_of_corruption', 'government_effectiveness', 'political_stability_and_absence_of_violence/terrorism', 'regulatory_quality', 'rule_of_law', 'voice_and_accountability']
wgi_baseline = wgi.groupby('country_code')[wgi_metrics].mean()
print("WGI Baseline Metrics:")
print(wgi_baseline.head())
```

# Synthetic Supplier Generation

```python
import random
import pandas as pd # Ensure pandas is imported for pd.isna

random.seed(42)

# ------ Lead Time Metrics ----------
AVG_SEMICON_CYCLE_DAYS = 70 # global average fab+packaging cycle/manufacturing time.

# Synthetic country-level baseline transit time to first U.S. entry point
# For ocean suppliers: foreign port -> U.S. port
# For CAN/MEX: supplier region -> U.S. land border
TRANSIT_TIME_BASE = {
    "ARE": 35,
    "AUS": 25,
    "BEL": 14,
    "BRA": 16,
    "CHN": 20,
    "DEU": 16,
    "FIN": 20,
    "FRA": 15,
    "GBR": 14,
    "HKG": 20,
    "IDN": 25,
    "IND": 35,
    "JPN": 14,
    "MYS": 24,
    "NLD": 14,
    "SGP": 26,
    "THA": 24
}

# Origin vessel turnaround times (in days) from US Port Call data
VESSEL_TURNAROUND_BASE = {
            "AUS": 1.4417,
            "CAN": 1.4569,
            "CHN": 0.6049,
            "DEU": 1.0174,
            "FRA": 0.8535,
            "GBR": 0.6812,
            "IDN": 0.9653,
            "JPN": 0.35,
            "SGP": 0.8521
        }
# These are treated as vessel turnaround, not full export dwell.
# Vessel turnaround: ship departure time - ship arrival time

# Asian countries should experience slightly higher U.S. port dwell because
# they are likely to enter through West Cost ports (more congestions)
ASIA_COUNTRIES = {"CHN", "HKG", "IDN", "JPN", "MYS", "THA", "SGP"}
# US Suppliers = No port dwell. No ocean transit. Only domestic logistics. E.g., trucking

synthetic_suppliers = []
supplier_id_counter = 1

# Iterate only through the country codes that are actually present in the combined_baselines DataFrame's index
for country_code in combined_baselines.index:
    country_baseline = combined_baselines.loc[country_code]
    # Generate 1 to n suppliers per country based on relevance in industry
    if country_code in tier_1:
      num_suppliers = random.randint(7,10)
    elif country_code in tier_2:
      num_suppliers = random.randint(2,4)
    else: # tier 3
      num_suppliers = random.randint(1,2)

    for i in range(num_suppliers):
        supplier_id = f"SUP_{country_code}_{supplier_id_counter}"

        # =========================
        # Lead Time Module
        # =========================

        #--------------------
        # 1. ORIGIN PORT DWELL - export side (supplier country i.e., how long product sits in own port before shipping)
        # -------------------


        if country_code == 'USA':
          origin_port_dwell = 0.0
        elif country_code in ['CAN', 'MEX']:
          origin_port_dwell = 0.0 # land border, not ocean export
        else:
          # vessel turnaround time from the observed US Port Calls data when available
          vessel_turnaround = VESSEL_TURNAROUND_BASE.get(country_code, None)
          if vessel_turnaround is None:
            vessel_turnaround = combined_baselines['median_days_in_port'].median()

          # LPI components for terminal/port-side friction of foreign suppliers
          customs = country_baseline.get('customs', 3.0)
          infrastructure = country_baseline.get('infrastructure', 3.0)
          logistics_competence = country_baseline.get('logistics_competence', 3.0)
          timeliness = country_baseline.get('timeliness', 3.0)

          # Weighted LPI quality score (1-5, higher = better)
          export_logistics_quality = (
              0.35 * customs + 0.25 * infrastructure + 0.20 * logistics_competence + 0.20 * timeliness)

          # Convert quality to terminal dwell
          # Better logistics -> lower dwell
          # Roughly maps strong countries to ~1 day and weaker ones to ~3+ days
          terminal_dwell = max(0.75, 4.5 - export_logistics_quality)
          # terminal dwell = Time container loaded onto ship - time container enters terminal

          # Full export-side dwell = vessel turnaround time + terminal dwell
          origin_port_dwell = (vessel_turnaround + terminal_dwell) * random.uniform(0.9, 1.1)

        # -------------------------
        # 2. Transit time to the US
        # -------------------------
        if country_code == "USA":
          transit_time = 0.0
        elif country_code == "CAN":
          transit_time = random.uniform(1.5, 3.0)
        elif country_code == "MEX":
          transit_time = random.uniform(2.5, 4.5)
        else:
          transit_base = TRANSIT_TIME_BASE.get(country_code, 22)
          transit_time = transit_base * random.uniform(0.9, 1.1)




        # ---------------------
        # 3. U.S. Port Dwell (seaborn imports)
        # Assumption: Washington, DC-based buyer

        if country_code in ["CAN", "MEX", "USA"]:
          us_port_dwell = 0.0
        else:
          if country_code in ASIA_COUNTRIES:
            us_port_dwell = random.uniform(3.25, 4.5)
          else:
            us_port_dwell = random.uniform(2.25, 3.5)

        # -----------------------
        # 4. land border crossing delay
        # Customs and querie crossing time at U.S. border
        # -------------------------
        if country_code == "CAN":
          border_delay = random.uniform(0.05, 0.25)
        elif country_code == "MEX":
          border_delay = random.uniform(0.25, 0.9)
        else:
          border_delay = 0.0

        # -------------------
        # 5. Inland Distribution
        # Assumption: Washington, DC-based buyer
        # Europe / Brazil / Middle East more likely East Cost entry -> shorter inland routing
        # Asia more likely West Cost entry -> longer inland routing to DC
        # Canada and Mexico treated as approximate inland shipment routes via freight truck or train
        # USA suppliers assumed to be Texas-based domestic suppliers shipping to DC
        # ---------------------
        if country_code == "USA":
          inland_distribution = random.uniform(2.75, 4.75) # Texas -> DC
        elif country_code == "CAN":
          inland_distribution = random.uniform(1.0, 2.5)
        elif country_code == "MEX":
          inland_distribution = random.uniform(2.0, 4.0) # captures nearby and farther regions
        elif country_code in ASIA_COUNTRIES:
          inland_distribution = random.uniform(4.0, 7.0) # likely West Coast entry -> DC
        else:
          inland_distribution = random.uniform(1.5, 4.0) # likely East/Gulf Cost entry -> DC

        # -----------------------
        # 6. Inspections / Customs Delay
        # Uses LPI customs as a simply proxy: lower customs score -> more possible inspection/clearance delay
        # ----------------------
        if country_code == "USA":
          inspection_delay = 0.0
        else:
          customs = country_baseline.get('customs', 3.0)
          # converts 1-5 customs score into extra delay
          # Good customs (4-5): very small delay
          # Weak customs (1-2): larger expected delay
          inspection_risk = max(0.0, # lower score better
                                min(1.0, (5.0 - customs) / 4.0))

          if country_code == "CAN":
            inspection_delay = inspection_risk * random.uniform(0.05, 0.35)
          elif country_code == "MEX":
            inspection_delay = inspection_risk * random.uniform(0.3, 1.0)
          else:
            inspection_delay = inspection_risk * random.uniform(0.2, 1.2)


        # ----------------
        # 7. TOTAL LEAD TIME MEAN
        # Includes semiconductor production/manufacturing cycle + logistics chains
        # ------------------
        if country_code == "USA" or country_code in tier_1:
            manufacturing_cycle = random.uniform(0.85, 0.95) * AVG_SEMICON_CYCLE_DAYS

        elif country_code in tier_2:
            manufacturing_cycle = random.uniform(0.95, 1.1) * AVG_SEMICON_CYCLE_DAYS

        else:  # tier_3
            manufacturing_cycle = random.uniform(1.05, 1.25) * AVG_SEMICON_CYCLE_DAYS

        lead_time_mean = (
            manufacturing_cycle +
            origin_port_dwell +
            transit_time +
            us_port_dwell +
            border_delay +
            inland_distribution +
            inspection_delay
        )

        # Prevent unrealistically low values as validation safety measure
        lead_time_mean = max(28, lead_time_mean)

        # --------------------

        # --------------------
        # Lead Time Volatility / Variance
        # --------------------

        # 1) Manufacturing cycle variability
        if country_code == "USA" or country_code in tier_1:
            manufacturing_cv = random.uniform(0.06, 0.10)
        elif country_code in tier_2:
            manufacturing_cv = random.uniform(0.08, 0.12)
        else:  # tier_3
            manufacturing_cv = random.uniform(0.10, 0.16)

        manufacturing_std = manufacturing_cycle * manufacturing_cv

        # 2) Origin port dwell variability
        if country_code in ["USA", "CAN", "MEX"]:
            origin_port_std = 0.0
        else:
            # Better logistics countries should have less dwell volatility
            export_efficiency = export_logistics_quality / 5.0  # 0 to 1
            origin_port_cv = max(0.10, 0.30 - 0.15 * export_efficiency)
            origin_port_std = origin_port_dwell * origin_port_cv

        # 3) Transit variability
        if country_code == "USA":
            transit_std = 0.0
        elif country_code in ["CAN", "MEX"]:
            transit_cv = random.uniform(0.06, 0.10)
            transit_std = transit_time * transit_cv
        elif country_code in ASIA_COUNTRIES:
            transit_cv = random.uniform(0.10, 0.16)
            transit_std = transit_time * transit_cv
        else:
            transit_cv = random.uniform(0.08, 0.13)
            transit_std = transit_time * transit_cv

        # 4) U.S. port dwell variability
        if country_code in ["USA", "CAN", "MEX"]:
            us_port_std = 0.0
        else:
            if country_code in ASIA_COUNTRIES:
                us_port_cv = random.uniform(0.18, 0.28)
            else:
                us_port_cv = random.uniform(0.14, 0.22)
            us_port_std = us_port_dwell * us_port_cv

        # 5) Border crossing variability
        if country_code == "CAN":
            border_cv = random.uniform(0.20, 0.35)
            border_std = border_delay * border_cv
        elif country_code == "MEX":
            border_cv = random.uniform(0.30, 0.50)
            border_std = border_delay * border_cv
        else:
            border_std = 0.0

        # 6) Inland distribution variability
        if country_code == "USA":
            inland_cv = random.uniform(0.10, 0.16)
        elif country_code == "CAN":
            inland_cv = random.uniform(0.10, 0.18)
        elif country_code == "MEX":
            inland_cv = random.uniform(0.12, 0.20)
        elif country_code in ASIA_COUNTRIES:
            inland_cv = random.uniform(0.12, 0.20)
        else:
            inland_cv = random.uniform(0.10, 0.18)

        inland_std = inland_distribution * inland_cv

        # 7) Inspection variability
        if inspection_delay == 0:
            inspection_std = 0.0
        else:
            inspection_cv = random.uniform(0.30, 0.50)
            inspection_std = inspection_delay * inspection_cv

        # Combine component variances assuming approximate independence
        lead_time_variance = (
            manufacturing_std ** 2 +
            origin_port_std ** 2 +
            transit_std ** 2 +
            us_port_std ** 2 +
            border_std ** 2 +
            inland_std ** 2 +
            inspection_std ** 2
        )

        lead_time_stddev = lead_time_variance ** 0.5 # take the sqrt to approximate std for lead time
        # --------------------

        # ==================================
        # --- Disruption Probability (WGI + LPI blended) ---
        # ===================================
        psat_score = country_baseline.get('political_stability_and_absence_of_violence/terrorism', 0)
        cc_score   = country_baseline.get('control_of_corruption', 0)

        # Normalize WGI (-2.5 to 2.5 → 0 to 1)
        normalized_psat = (psat_score + 2.5) / 5.0
        normalized_cc   = (cc_score + 2.5) / 5.0

        wgi_risk = 1 - ((normalized_psat + normalized_cc) / 2)

        # LPI operational risk (1–5 → 0–1 risk)
        infrastructure = country_baseline.get('infrastructure', 3.0)
        logistics_competence = country_baseline.get('logistics_competence', 3.0)

        normalized_infra = 1 - (infrastructure / 5.0)
        normalized_comp  = 1 - (logistics_competence / 5.0)

        lpi_risk = (normalized_infra + normalized_comp) / 2

        # Blend institutional + operational risk
        disruption_prob_base = 0.6 * wgi_risk + 0.4 * lpi_risk

        # Add proportional noise
        noise = random.uniform(-0.1, 0.1) * disruption_prob_base
        disruption_probability = disruption_prob_base + noise

        # Clamp
        disruption_probability = max(0.01, min(0.99, disruption_probability))
        # ----- End Disruption Probability --------

        # ------------------------
        # Compliance Eligibility: How likely supplier is to meet corporate, regulatory, ethical, and trade-compliance requirements
        # How clean, compliant, low-risk partner is from a governance and regulatory standpoint
        # Directly related to WGI metrics such as 'control_of_corruption','regulatory_quality', 'rule of law', 'political stablity', 'lpi customs' with randomness.
        # --- Compliance Eligibility (WGI + LPI Customs blended) ---

        cc_score   = country_baseline.get('control_of_corruption', 0)
        rq_score   = country_baseline.get('regulatory_quality', 0)
        rol_score  = country_baseline.get('rule_of_law', 0)
        psat_score = country_baseline.get('political_stability_and_absence_of_violence/terrorism', 0)

        customs = country_baseline.get('customs', 3.0)  # LPI Customs (1–5)

        # Normalize WGI (-2.5 → +2.5 mapped to 0–1)
        norm_cc   = (cc_score + 2.5) / 5.0
        norm_rq   = (rq_score + 2.5) / 5.0
        norm_rol  = (rol_score + 2.5) / 5.0
        norm_psav = (psat_score + 2.5) / 5.0

        # Normalize LPI Customs (1–5 → 0–1)
        norm_customs = customs / 5.0

        # Weighted compliance model
        compliance_base = (
          0.40 * norm_cc +
          0.25 * norm_rq +
          0.20 * norm_rol +
          0.05 * norm_psav +
          0.10 * norm_customs
            )

        # Add proportional noise (±5–10%)
        noise = random.uniform(-0.10, 0.10) * compliance_base
        compliance_eligibility = compliance_base + noise

        # Clamp to 5%–95%
        compliance_eligibility = max(0.05, min(0.95, compliance_eligibility))





        # -----------------------

        # --------------
        # --- Logistics Reliability (weighted LPI model) ---

        customs = country_baseline.get('customs', 3.0)
        infrastructure = country_baseline.get('infrastructure', 3.0)
        logistics_competence = country_baseline.get('logistics_competence', 3.0)
        tracking = country_baseline.get('tracking', 3.0)
        timeliness = country_baseline.get('timeliness', 3.0)

        # Weighted average of LPI pillars
        weighted_lpi = (
            0.30 * timeliness +
            0.25 * logistics_competence +
            0.20 * infrastructure +
            0.15 * tracking +
            0.10 * customs
        )

        # Convert 1–5 LPI scale → 0–1 reliability
        base_reliability = weighted_lpi / 5.0

        # Add proportional noise (±5%)
        noise = random.uniform(-0.05, 0.05) * base_reliability
        logistics_reliability = base_reliability + noise

        # Clamp to 10%–99%
        logistics_reliability = max(0.10, min(0.99, logistics_reliability))

        # -----------------






        synthetic_suppliers.append({
            'country_code': country_code,
            'supplier_id': supplier_id,
            'lead_time_mean': lead_time_mean,
            'lead_time_stddev': lead_time_stddev,
            'lead_time_variance': lead_time_variance,
            'disruption_probability': disruption_probability,
            'compliance_eligibility': compliance_eligibility,
            'logistics_reliability': logistics_reliability,
        })
        supplier_id_counter += 1

synthetic_suppliers_df = pd.DataFrame(synthetic_suppliers)

print("Synthetic Supplier Profiles (first 5 rows):")
print(synthetic_suppliers_df.head())
```
